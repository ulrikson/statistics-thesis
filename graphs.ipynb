```python
import pandas as pd
from datetime import datetime, timedelta, date

import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/final.csv')
df['Exchange.Date'] = df['Exchange.Date'].apply(lambda x: date(1900, 1, 1) + timedelta(int(x)))

df.head()
```

```python
plt.figure(figsize=(12, 5))
plt.rcParams.update({'font.size': 18})

plt.plot(df['Exchange.Date'], df["Close"], label="Index (kr)", color="black", linewidth=1)

plt.xlabel('Datum')
plt.ylabel('IndexvÃ¤rde') 
# plt.title('OMXSSC i kronor')

plt.savefig('img/index.png')
```

```python
plt.figure(figsize=(12, 5))
plt.rcParams.update({'font.size': 18})

plt.plot(df['Exchange.Date'], df["logreturns"], label="Logaritmerad avkastning", color="black", linewidth=0.5)

plt.xlabel('Datum')
plt.ylabel('Logartimerad avkastning') 
# plt.title('OMXSSC som logaritmerad avkastning')

plt.savefig('img/logreturns.png')
```
